```python
import pandas as pd
```

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
import seaborn as sb
sb.set()
```

```python
url='https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
```

```python
df['Sex'].unique()
encoded_dict={"male":0,"female":1}
df["Sex"]=df["Sex"].map(encoded_dict)
```

```python
sb.catplot( "Sex", "Age", data=df, kind="box")
plt.show()
```

```python
sb.distplot(df['Age'])
sb.distplot(df['Sex']);
plt.title("Seaborn distplot")
plt.show()
```

```python
graph=sb.scatterplot(x="Age",y="Survived",data=df)
```
